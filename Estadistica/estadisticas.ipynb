```python
import pandas as pd


dataRead = pd.read_excel('Estadistica.xlsx', sheet_name='Cuantitativos_Discretos')


dataRead.info()
```

```python
# Frecuancia absoluta
frecuencia = dataRead['Talla_Zapatos'].value_counts().reset_index()

frecuencia.columns = ['Talla_Zapatos', 'Frecuencia_Absoluta']

frecuencia.head(20)
```

```python
# Frecuancia absoluta acumulada
frecuencia['Frecuencia_Acumulada_Absoluta'] = frecuencia['Frecuencia_Absoluta'].cumsum()

frecuencia.head(14)
```

```python
# Frecuencia Relativa
suma_absoluta = frecuencia['Frecuencia_Absoluta'].sum()

frecuencia['Frecuencia_Relativa'] = frecuencia['Frecuencia_Absoluta'] / suma_absoluta * 100

frecuencia.head(14)
```

```python
# Frecuencia Relativa Acumulada
frecuencia['Frecuencia_Acumulada_Relativa'] = frecuencia['Frecuencia_Relativa'].cumsum()

frecuencia.head(14)
```

```python
talla_zapatos = dataRead.groupby('Talla_Zapatos').agg({'Talla_Zapatos': 'count'})
talla_zapatos.plot(kind='bar', title='Talla_Zapatos', ylabel='Cantidad', color='blue')
```

```python
import numpy as np
import pandas as pd

dataSet = pd.read_excel("Estadistica.xlsx", sheet_name='Cuantitativos_Continuos')

num_intervalos = int(1 + 3.322 * np.log10(len(dataSet)))

dataSet['Intervalos'] = pd.cut(dataSet['Peso_Estudiantes'], bins=num_intervalos)

dataSet
```

```python
frecuencia = dataSet['Intervalos'].value_counts().sort_index().reset_index()
frecuencia.columns = ['Intervalos', 'Frecuencia_Absoluta']

frecuencia['Frecuencia_Acumulada_Absoluta'] = frecuencia['Frecuencia_Absoluta'].cumsum()

frecuencia
```

```python
suma_absoluta = frecuencia['Frecuencia_Absoluta'].sum()
frecuencia['Frecuencia_Relativa'] = frecuencia['Frecuencia_Absoluta'] / suma_absoluta * 100

frecuencia['Frecuencia_Acumulada_Relativa'] = frecuencia['Frecuencia_Relativa'].cumsum()

frecuencia
```

```python
intervalos = dataSet.groupby('Intervalos').agg({'Intervalos': 'count'})
intervalos.plot(kind='bar', title='Intervalos', ylabel='Cantidad', color='purple')
```

```python
import pandas as pd
datasetC = pd.read_excel('Estadistica.xlsx', sheet_name = 'Cualitativos')

# Calcula la frecuencia absoluta utilizando el método value_counts() de Pandas
frecuencia = datasetC['Nivel_Escolar'].value_counts().reset_index()
frecuencia.columns = ['Nivel_Escolar', 'Frecuencia_Absoluta']

# Calcular la frecuencia absoluta acumulada
frecuencia['Frecuencia_Acumulada_Absoluta'] = frecuencia['Frecuencia_Absoluta'].cumsum()

frecuencia
```

```python

# Calcula la frecuencia relativa (frecuencia absoluta / suma de todas las frecuencias absolutas)
suma_absoluta = frecuencia["Frecuencia_Absoluta"].sum()
frecuencia["Frecuencia_Relativa"] = frecuencia["Frecuencia_Absoluta"] / suma_absoluta * 100

# Calcula la frecuencia relativa acumulada
frecuencia['Frecuencia_Acumulada_Relativa'] = frecuencia['Frecuencia_Relativa'].cumsum()

frecuencia
```

```python
nivel_Escolar = datasetC.groupby('Nivel_Escolar').agg({'Nivel_Escolar': 'count'})
nivel_Escolar.plot(kind='bar', title='Nivel_Escolar', ylabel='Cantidad', color='orange')
```
