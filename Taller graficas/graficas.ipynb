```python
import pandas as pd


data = pd.read_excel('datosnbanco.xlsx')

data.info()
```

```python
del data["Unnamed: 0"]

data.info()
```

```python
data["job"].unique()
```

```python
data.groupby('job')['age'].count()
```

```python
data["job"] = data["job"].str.replace("Retired", "retired", regex=False)

data["job"] = data["job"].str.replace("Services", "services", regex=False)

data["job"].unique()
```

```python
agesBar = data.groupby('job')['age'].max()

agesBar
```

```python
agesCount = data.groupby('job')['age'].count()

agesCount
```

```python
agesSum = data.groupby('job')['age'].sum()

agesSum
```

```python
agesMin = data.groupby('job')['age'].min()

agesMin
```

```python
agesPie = data.groupby('job')['age'].mean()

agesPie
```

```python
import matplotlib.pyplot as plt

jobs = agesBar.index
max_ages = agesBar.values

fig, ax = plt.subplots()

ax.bar(jobs, max_ages)

plt.xticks(rotation=45, ha='right')

plt.xlabel('Trabajo')
plt.ylabel('Edad mxima')
plt.title('Edad mxima por trabajo')

plt.show()
```

```python
jobs = agesCount.index
count_ages = agesCount.values

fig, ax = plt.subplots()

ax.plot(jobs, count_ages)

plt.xticks(rotation=45, ha='right')

plt.xlabel('Trabajo')
plt.ylabel('Total de edades')
plt.title('Conteo de edades por trabajo')

plt.show()
```

```python
jobs = agesSum.index
sum_ages = agesSum.values

fig, ax = plt.subplots()

ax.barh(jobs, sum_ages)

plt.xticks(rotation=45, ha='right')

plt.xlabel('Trabajo')
plt.ylabel('Suma de edades')
plt.title('Suma de edades por trabajo')

plt.show()
```

```python
jobs = agesMin.index
min_ages = agesMin.values

fig, ax = plt.subplots()

ax.fill_between(jobs, min_ages)

plt.xticks(rotation=45, ha='right')

plt.xlabel('Trabajo')
plt.ylabel('Edad mnima')
plt.title('Edad mnima por trabajo')

plt.show()
```

```python
jobs = agesPie.index
min_ages = agesPie.values

fig, ax = plt.subplots()

ax.scatter(jobs, min_ages)

plt.xticks(rotation=45, ha='right')

plt.xlabel('Trabajo')
plt.ylabel('Edad mnima')
plt.title('Edad mnima por trabajo')

plt.show()
```
